# 이미지 저장하기

```python
img = cv2.imread('./fig/PD_MRI.jpeg', cv2.IMREAD_GRAYSCALE)

# cv2.resize(src=source,
#     dsize=resize_shape[, dst, fx, fy, interpolation])
reduced_mri = cv2.resize(img,
                         (300, 300),
                         interpolation=cv2.INTER_AREA)
cv2.imwrite('my_brain.jpeg', reduced_mri)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('image read failed')
    sys.exit('read failed')
    
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)

reduced_img = cv2.resize(img,(100,100), interpolation=cv2.INTER_AREA)
cv2.imwrite('./fig/my_puppy.png', reduced_img)

# cv2.waitKey(3000) # 3초 후 넘어감.
while True:
    kw = cv2.waitKey()
    if kw == 27 or kw == 113: # 27-esc // 13-enter // 113-'q'
        break
    
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
imgBGR = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)

if imgBGR is None:
    print('image read failed')
    sys.exit()
    
cv2.imshow('image', imgBGR)

while True:
    kw = cv2.waitKey()
    if kw == 113 or kw==27:
        break
        
cv2.destroyAllWindows()
```

```python
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
imgGray = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)


fig, axes = plt.subplots(1,3,figsize=(12,12))
axes[0].imshow(imgRGB), axes[0].axis('off')
axes[1].imshow(imgBGR), axes[1].axis('off')
axes[2].imshow(imgGray, cmap='gray'), axes[2].axis('off')

# plt.figure(figsize=(8,8))
# plt.imshow(imgRGB)
# plt.axis('off')
plt.show()
```
